```python
import pandas as pd
import json

file_list = [
    'response_1713698731707 - DOLNOŚLĄSKIE.json',
    'response_1713698773810 - ŁÓDZKIE.json',
    'response_1713698832489 - LUBELSKIE.json',
    'response_1713698866840 - LUBUSKIE.json',
    'response_1713698891060 - MAZOWIECKIE.json',
    'response_1713698946109 - OPOLSKIE.json',
    'response_1713698973885 - PODKARPACKIE.json',
    'response_1713698993320 - PODLASKIE.json',
    'response_1713699019535 - POMORSKIE.json',
    'response_1713699968171 - ŚLĄSKIE.json',
    'response_1713699050048 - ŚWIĘTOKRZYSKIE.json',
    'response_1713699071764 - WARMIŃSKO-MAZURSKIE.json',
    'response_1713699104738 - WIELKOPOLSKIE.json',
    'response_1713699810178 - ZACHODNIOPOMORSKIE.json',
    'response_1713699863462 - KUJAWSKO-POMORSKIE.json',
    'response_1713699918895 - MAŁOPOLSKIE.json'
    ]

dfs = []
for file in file_list:
    with open(file, 'r', encoding='utf-8') as f:
        json_file = json.load(f)
        df = pd.DataFrame(json_file['Przekroczenia'])
        dfs.append(df)

df = pd.concat(dfs, ignore_index=True)

df['Stanowisko pomiarowe'] = df['Stanowisko pomiarowe'].str.split('-', expand=True)[0]

df['Data i godzina'] = pd.to_datetime(df['Data i godzina'], format='%Y-%m-%d %H:%M')

df.head()
```

```python
json_stations = json.load(open('response_1711986980486 - list of stations.json', encoding='utf-8'))

json_stations['Lista stacji pomiarowych']

df_stations = pd.DataFrame(json_stations['Lista stacji pomiarowych'])

df_stations.head()
```

```python
df3 = pd.merge(df,df_stations,how='left',left_on='Stanowisko pomiarowe',right_on='Kod stacji')

df3.head()
```

```python
df3 = df3.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
df3.to_csv('data_clean.csv', index=False, lineterminator='\n', encoding='utf-8', escapechar='\\')
```
